```python
# 起始数（十六进制）
start = '0'
# start = '2024'

# 增量（十进制）
delta = 1

# 输出个数
count = 50

# 输出位数，-1表示不指定位数
length = 8

# 是否打印起始数
print_start = True

# 是否大写
print_upper = True

# 开头是否打印“0x”
print_0x = False

# 结尾是否打印“h”
print_h = False
```

```python
def convert_decimal_to_hexadecimal(decimal):
    hexadecimal = hex(decimal).replace("0x", "")
    
    if print_upper:
        hexadecimal = hexadecimal.upper()
    
    if length != -1:
        count_zero = length - len(hexadecimal) if length - len(hexadecimal) > 0 else 0
        hexadecimal = "0" * count_zero + hexadecimal
    
    if print_0x:
        hexadecimal = "0x" + hexadecimal
    
    if print_h:
        hexadecimal = hexadecimal + "h"
    
    return hexadecimal
```

```python
def print_order_array():
    number = int(start, 16)
    
    if print_start:
        print(convert_decimal_to_hexadecimal(number))
    
    for i in range(1, count):
        number = number + delta
        print(convert_decimal_to_hexadecimal(number))

print_order_array()
```

```python
# 起始数（十六进制）
base = '40D4B9'
# base = '0'

# 增量（十六进制）
deltas = [
'00',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'01',
'04',
'04',
'04',
'04',
'04',
'04',
'02',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'04',
'03',
'CC',
'CC'
]
```

```python
def print_disorder_array():
    for delta in deltas:
        number = int(base, 16) + int(delta, 16) * 4
        # number = int(base, 16) + int(delta, 16)
        print(convert_decimal_to_hexadecimal(number))

print_disorder_array()
```
